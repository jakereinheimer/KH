```python
#PROJECTION
import yt
import os
import cv2

nfiles=60
folder='data9'
location='D:/sim_data/cluster_data/kh_data/'+folder
dimensions=['z']
types=["rho",'press']

units_override = {
    "length_unit": (1.0, "Mpc"),
    "time_unit": (1.0, "Myr"),
}

#PROJECTION

def make_plots():
    for dim in dimensions:
        for plot_type in types:
            contents=os.listdir(location)
            if (f'Projection,{dim},{plot_type}') not in contents:
                os.mkdir(location+'/'+f'Projection,{dim},{plot_type}')
                
                
    for i in range(nfiles):
        ind='%05d' % (i)

        for dim in dimensions:
            for plot_type in types:
                ds0=yt.load(location+f'/kh-custom.out1.{ind}.athdf',units_override=units_override)
                p=yt.ProjectionPlot(ds0,dim,plot_type)
                '''
                if plot_type == "press":
                    p.set_zlim('press',zmin=(10), zmax=(10**3))
                if plot_type == "rho":
                    p.set_zlim('rho',zmin=(10), zmax=(10**3))
                else:
                    p.set_zlim(plot_type,zmin=(10), zmax=(10**3))
                '''
                p.set_zlim(plot_type,zmax=(10**3))
                p.set_cmap(plot_type,"seismic")
                p.save(location+f'/Projection,{dim},{plot_type}/kh-custom.out1.{ind}.png')

def make_videos():
    #create videos
    for dim in dimensions:
        for plot_type in types:
            specific_loc=location+f'/projection,{dim},{plot_type}'
            pics=os.listdir(specific_loc)
            img_array=[]
            for pic in pics:
                img=cv2.imread(specific_loc+'/'+pic)
                height,width,layers=img.shape
                size=(width,height)
                img_array.append(img)

            out=cv2.VideoWriter(specific_loc+"/video.mp4",cv2.VideoWriter_fourcc(*'DIVX'), 15, size)

            for i in range(len(img_array)):
                out.write(img_array[i])
            out.release()
            
make_plots()
make_videos()
```

```python
#SLICE
import yt
import os
import cv2
import sympy

nfiles=7
folder='data21'
location='D:/sim_data/cluster_data/kh_data/'+folder
dimensions=['z','x','y']
types=["rho","press","vel1","vel2","vel3"]

units_override = {
    "length_unit": (1.0, "Mpc"),
    "time_unit": (1.0, "Myr"),
}

#SLICE

def make_plots():
    
    for dim in dimensions:
        for plot_type in types:
            contents=os.listdir(location)
            if (f'slice,{dim},{plot_type}') not in contents:
                os.mkdir(location+'/'+f'slice,{dim},{plot_type}')

            
    for i in range(nfiles):
        ind='%05d' % (i)
            
        for dim in dimensions:
            for plot_type in types:
                ds0=yt.load(location+f'/kh-custom.out1.{ind}.athdf',units_override=units_override)
                p=yt.SlicePlot(ds0,dim,plot_type,center=[0,0,0])
                #p.set_zlim(plot_type,zmax=(10**3))
                p.set_cmap(plot_type,"seismic")
                p.save(location+f'/slice,{dim},{plot_type}/kh-custom.out1.{ind}.png')

def make_videos():
    #create videos
    for dim in dimensions:
        for plot_type in types:
            specific_loc=location+f'/slice,{dim},{plot_type}'
            pics=os.listdir(specific_loc)
            img_array=[]
            for pic in pics:
                img=cv2.imread(specific_loc+'/'+pic)
                height,width,layers=img.shape
                size=(width,height)
                img_array.append(img)

            out=cv2.VideoWriter(specific_loc+"/video.mp4",cv2.VideoWriter_fourcc(*'DIVX'), 15, size)

            for i in range(len(img_array)):
                out.write(img_array[i])
            out.release()
            
make_plots()
make_videos()
```
