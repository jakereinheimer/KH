```python
#PROJECTION
import yt
import os
import cv2

nfiles=60
folder='data9'
location='D:/sim_data/cluster_data/kh_data/'+folder
dimensions=['z']
types=["rho",'press']

units_override = {
    "length_unit": (1.0, "Mpc"),
    "time_unit": (1.0, "Myr"),
}

#PROJECTION

def make_plots():
    for dim in dimensions:
        for plot_type in types:
            os.mkdir(location+'/'+f'Projection,{dim},{plot_type}')
    for i in range(nfiles):
        if i <10:
            ind=str(i)
            ind='0'+ind
        else:
            ind=str(i)

        for dim in dimensions:
            for plot_type in types:
                ds0=yt.load(location+f'/kh-custom.out1.000{ind}.athdf',units_override=units_override)
                p=yt.ProjectionPlot(ds0,dim,plot_type)
                '''
                if plot_type == "press":
                    p.set_zlim('press',zmin=(10), zmax=(10**3))
                if plot_type == "rho":
                    p.set_zlim('rho',zmin=(10), zmax=(10**3))
                else:
                    p.set_zlim(plot_type,zmin=(10), zmax=(10**3))
                '''
                p.set_zlim(plot_type,zmax=(10**3))
                p.set_cmap(plot_type,"seismic")
                p.save(location+f'/Projection,{dim},{plot_type}/kh-custom.out1.000{ind}.png')

def make_videos():
    #create videos
    for dim in dimensions:
        for plot_type in types:
            specific_loc=location+f'/projection,{dim},{plot_type}'
            pics=os.listdir(specific_loc)
            img_array=[]
            for pic in pics:
                img=cv2.imread(specific_loc+'/'+pic)
                height,width,layers=img.shape
                size=(width,height)
                img_array.append(img)

            out=cv2.VideoWriter(specific_loc+"/video.mp4",cv2.VideoWriter_fourcc(*'DIVX'), 15, size)

            for i in range(len(img_array)):
                out.write(img_array[i])
            out.release()
            
make_plots()
make_videos()
```

```python
#SLICE
import yt
import os
import cv2
import sympy

nfiles=10
folder='data19'
location='D:/sim_data/cluster_data/kh_data/'+folder
dimensions=['z','x','y']
types=["rho","press","vel1","vel2","vel3"]

units_override = {
    "length_unit": (1.0, "Mpc"),
    "time_unit": (1.0, "Myr"),
}

#SLICE

def make_plots():
    for dim in dimensions:
        for plot_type in types:
            os.mkdir(location+'/'+f'slice,{dim},{plot_type}')
    for i in range(nfiles):
        if i <10:
            ind=str(i)
            ind='0'+ind
        else:
            ind=str(i)

        for dim in dimensions:
            for plot_type in types:
                ds0=yt.load(location+f'/kh-custom.out1.000{ind}.athdf',units_override=units_override)
                p=yt.SlicePlot(ds0,dim,plot_type,center=[0,0,0])
                #p.set_zlim(plot_type,zmax=(10**3))
                p.set_cmap(plot_type,"seismic")
                p.save(location+f'/slice,{dim},{plot_type}/kh-custom.out1.000{ind}.png')

def make_videos():
    #create videos
    for dim in dimensions:
        for plot_type in types:
            specific_loc=location+f'/slice,{dim},{plot_type}'
            pics=os.listdir(specific_loc)
            img_array=[]
            for pic in pics:
                img=cv2.imread(specific_loc+'/'+pic)
                height,width,layers=img.shape
                size=(width,height)
                img_array.append(img)

            out=cv2.VideoWriter(specific_loc+"/video.mp4",cv2.VideoWriter_fourcc(*'DIVX'), 15, size)

            for i in range(len(img_array)):
                out.write(img_array[i])
            out.release()
            
make_plots()
make_videos()
```

```python
import numpy as np
import yt

colormap="seismic"

rho_naut=100
total_n=60

for i in range(total_n-1):
    i+=1
    
    ds=yt.load("D:/sim_data/cluster_data/winter_break_data/data1/kh-custom.out1.000%.2i.athdf"%i)

    def rho_ratio(field,data):
        return data["rho"]/rho_naut

    ds.add_field(
        name=("gas","rho_ratio"),
        function=rho_ratio,
        sampling_type="local",
        units="auto",
        force_override=True
    )

    slc=yt.SlicePlot(ds,"z",("gas","rho_ratio"))

    slc.set_cmap(("gas","rho_ratio"),colormap)

    slc.save("D:/sim_data/cluster_data/winter_break_data/data1/videos/%.2f.png"%i)
```
